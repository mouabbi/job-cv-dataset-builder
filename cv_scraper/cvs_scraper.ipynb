```python
import pandas as pd
import matplotlib.pyplot as plt
from utils.cv_extractor import load_domains,filter_cvs_by_domain_and_category
# Load the dataset
df = pd.read_csv('cvs.csv')
```

```python
# Inspect the first few rows
df.head()
```

```python
# Count the number of resumes in each unique category
category_counts = df['Category'].value_counts()

# Plotting the distribution of unique categories
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Resumes by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Resumes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
